```python
import pandas as pd
df1 = pd.DataFrame({'fruits':['apple', 'mango', 'banana'],
                    'price':['100', '600', '60']})
display('the first dataframe')
display(df1)

df2=pd.DataFrame({'vegetables':['Onion', 'Brinjal', 'cabbage'],
                  'price':['60', '50', '30']})
display('the second dataframe')
display(df2)

display('the merged dataframe')
pd.merge(df1, df2, on = 'price', how= 'inner')
#pd.merge(df1, df2, on = 'price', how = 'outer')
#how --> left, right, inner, outer but default outer
```

```python
import pandas as pd
df1 = pd.DataFrame({'fruits':['apple', 'mango', 'banana'],
                    'price':['100', '600', '60']})
display('the first dataframe')
display(df1)

df2=pd.DataFrame({'vegetables':['Onion', 'Brinjal', 'cabbage'],
                  'price':['60', '50', '30']})
display('the second dataframe')
display(df2)

display('the merged dataframe')
pd.merge(df1, df2, on = 'price', how= 'inner')
pd.merge(df1, df2, on = 'price', how = 'outer')
```
